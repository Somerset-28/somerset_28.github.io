## Add New Entry Fields

### Subtask:
Implement input fields (e.g., text widgets for each column) and a button to allow the user to input data for a new row and add it to the DataFrame.


**Reasoning**:
First, I need to import the `ipywidgets` library as `widgets` to create interactive input fields and a button, as instructed.



```python
import ipywidgets as widgets
from IPython.display import display, HTML

print("ipywidgets imported successfully.")
```

**Reasoning**:
I need to create a dictionary of input widgets, one for each column in the `df` DataFrame, to allow users to input data for a new row. I will iterate through `df.columns` to create `widgets.Text` widgets and store them in a dictionary for easy access.



```python
input_widgets = {}
for col in df.columns:
    input_widgets[col] = widgets.Text(description=col)

print("Input widgets created for all columns.")
```

**Reasoning**:
Now that input widgets for each column are created, the next step is to create a button widget that will trigger the action of adding a new row to the DataFrame. This button will be linked to a function that processes the input from the widgets.



```python
add_row_button = widgets.Button(description='Add Row')

print("Add Row button created.")
```

**Reasoning**:
I need to define a function that collects values from the input widgets, creates a new row, appends it to the `df` DataFrame, clears the input widgets, and updates the `qgrid` display. This function will be linked to the `add_row_button`.



```python
import pandas as pd

def add_row(b):
    global df, qgrid_widget
    new_row_data = {col: widget.value for col, widget in input_widgets.items()}

    # Ensure data types match existing DataFrame columns if possible
    # For simplicity, we'll try to convert to the existing column types
    for col, value in new_row_data.items():
        if col in df.columns:
            try:
                # Attempt to convert to the column's dtype
                new_row_data[col] = pd.Series([value], dtype=df[col].dtype).iloc[0]
            except:
                # If conversion fails, keep as string or handle appropriately
                pass

    # Convert the new row data to a Series for proper appending
    new_row_series = pd.Series(new_row_data)

    df = pd.concat([df, pd.DataFrame([new_row_series])], ignore_index=True)

    # Clear input widgets
    for widget in input_widgets.values():
        widget.value = ''

    # Update the qgrid widget if it's already displayed
    if qgrid_widget is not None:
        qgrid_widget.df = df

    print("New row added and DataFrame updated.")
```

**Reasoning**:
Now that the `add_row` function is defined, I need to attach it to the `add_row_button`'s `on_click` event so that the function is executed when the button is pressed. Afterward, I will arrange and display all the input widgets and the button using `widgets.VBox` for better organization in the output.



```python
add_row_button.on_click(add_row)

# Arrange and display widgets
input_widgets_list = list(input_widgets.values())

# Use a VBox to stack the input widgets vertically, and then the button
interface = widgets.VBox(input_widgets_list + [add_row_button])
display(interface)

print("Input interface displayed.")
```

## Save Changes and Display Updated Table

### Subtask:
Provide functionality to process the edits and new entries, update the underlying DataFrame, and then save and display the modified DataFrame to reflect these changes.


**Reasoning**:
I need to create a button for saving the changes made to the DataFrame, as per the first instruction. This button will trigger the save functionality.



```python
import ipywidgets as widgets
from IPython.display import display

save_button = widgets.Button(description='Save Changes')
print("Save Changes button created.")
```

**Reasoning**:
Now that the `save_button` has been created, I need to define the `save_changes` function. This function will be responsible for retrieving the updated DataFrame from the `qgrid_widget`, saving it to an Excel file, and then displaying the head of the DataFrame to confirm the changes, as instructed.



```python
import pandas as pd

def save_changes(b):
    global df, qgrid_widget

    # Get the changed DataFrame from qgrid_widget
    # qgrid_widget.get_changed_df() returns the DataFrame with the edits
    df = qgrid_widget.get_changed_df()

    # Save the updated DataFrame to a new Excel file
    output_filename = 'updated_rotation_automated.xlsx'
    df.to_excel(output_filename, index=False)

    print(f"DataFrame saved to '{output_filename}' and updated.")

    # Display the first few rows of the updated DataFrame
    display(df.head())

print("save_changes function defined.")
```

**Reasoning**:
I need to attach the `save_changes` function to the `save_button`'s `on_click` event and then display the button, as instructed.



```python
save_button.on_click(save_changes)
display(save_button)

print("Save Changes button displayed and linked.")
```
